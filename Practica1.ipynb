**1) Adivina numero.**

Admin introduce un número y lo almacena en una variable.

Pide al usuario que adivine un número. Si el usuario se equivoca, se le indica si el numero es mayor o menor. 

El programa termina cuando el usuario ha adivinado el número.

**Ejemplo**:

introduzca numero: ***

adivine numero: 5

Intento erróneo. El número a adivinar es mayor.

adivine numero: 10

Intento erróneo. El número a adivinar es mayor.

adivine numero: 20.

Ha acertado el número!

```
import getpass

numero_adivinar = int(getpass.getpass('Introduzca el número para jugar: '))

while True:
  numero = int(input('Adivine número: '))
  if numero > numero_adivinar:
    print(f'Intento erróneo. El número a adivinar es menor que {numero}')
  elif numero < numero_adivinar:
    print(f'Intento erróneo. El número a adivinar es mayor que {numero}')
  else:
    print('Ha acertado el número!')
    break
```

**2) Adivina número con intentos finitos.**

Admin introduce un número a adivinar y un número máximo de intentos para adivinarlo.

Pide al usuario que adivine un número. Si el usuario se equivoca, se le indica si el numero es mayor o menor y el número de intentos restantes.

El programa termina cuando el usuario ha adivinado el número o si se supera el máximo número de intentos. Al final independientemente de haber adivinado o no, muestra los intentos insertados.

Ejemplo:

introduzca numero: ***

introduzca intentos permitidos: 10

adivine numero: 5

Intento erróneo. El número a adivinar es mayor. 9 intentos disponibles.

adivine numero: 10

Intento erróneo. El número a adivinar es mayor. 8 intentos disponibles.

adivine numero: 20.

Ha acertado el número! Los intentos fueron los siguientes: 5, 10, 20. 

```
import getpass

numero_adivinar = int(getpass.getpass('Introduzca el número para jugar: '))
numero_intentos = int(input('Introduzca el número de intentos permitidos: '))
lista = []

while True:
  if numero_intentos == 0:
    lista_f = (', ').join(lista)
    print(f'Has perdido! Los intentos fueron los siguientes: {lista_f}.')
    break
  else:
    numero = int(input('Adivine número: '))
    lista.append(str(numero))
    if numero > numero_adivinar:
      numero_intentos -= 1
      print(f'Intento erróneo. El número a adivinar es menor que {numero}. {numero_intentos} intentos disponibles.')
    elif numero < numero_adivinar:
      numero_intentos -= 1
      print(f'Intento erróneo. El número a adivinar es menor que {numero}. {numero_intentos} intentos disponibles.')
    else:
      lista_f = (', ').join(lista)
      print(f'Ha acertado el número! Los intentos fueron los siguientes: {lista_f}.')
      break
```

**4) Adivina frase.**

Admin introduce una frase a adivinar y un número máximo de intentos para adivinarlo.

Pide al usuario adivinar el número de palabras (ej1) en la frase.  Si el usuario se equivoca, se le indica si el numero es mayor o menor y el número de intentos restantes.

Pide al usuario si desea adivinar letras o palabras:
 

*   Letras: Se pide la letra. Muestra al usuario las letras que se van adivinando en la frase. Si se equivoca se resta 1 intento.

*   Adivinar frase: Si se equivoca se resta 1 intento.


El programa termina cuando el usuario ha adivinado la frase o si se supera el máximo número de intentos. Al final independientemente de haber adivinado o no, muestra los intentos insertados por tipo letra o palabra.

Los intentos de frase fueron los siguientes: el gato come.

```
import getpass

frase_adivinar = getpass.getpass('Introduzca la frase para jugar: ')
numero_intentos = int(input('Introduzca el número de intentos permitidos: '))

frase_separada = (frase_adivinar.lower()).split()

acertar_palabras = False
lista_intentos_num = []
lista_intentos_letras = []
lista_intentos_palabras = []
lista_intentos_frases = []
frase_pista = []

for palabra in frase_separada:
  frase_pista.extend(['*'] * len(palabra))
  if palabra == frase_separada[-1]: break 
  else: frase_pista.append(' ')

while True:
  if numero_intentos == 0:
    print('Has perdido!')
    lin_f = (', ').join(lista_intentos_num)
    lil_f = (', ').join(lista_intentos_letras)
    lip_f = (', ').join(lista_intentos_palabras)
    lif_f = (', ').join(lista_intentos_frases)
    print(f'Los intentos numero de letras fueron los siguientes: {lin_f}.')
    print(f'Los intentos de letras fueron los siguientes: {lil_f}.')
    print(f'Los intentos de palabras fueron los siguientes: {lip_f}.')
    print(f'Los intentos de frases fueron los siguientes: {lif_f}.')
    break
  else:
    if not acertar_palabras:
      numero = int(input('Adivine el número de palabras: '))
      lista_intentos_num.append(str(numero))
      if numero > len(frase_separada):
        numero_intentos -= 1
        print(f'Intento erróneo. La frase tiene menos de {numero} palabras. {numero_intentos} intentos disponibles.')
      elif numero < len(frase_separada):
        numero_intentos -= 1
        print(f'Intento erróneo. La frase tiene más de {numero} palabras. {numero_intentos} intentos disponibles.')
      else:
        acertar_palabras = True
        print(f'Ha acertado el número de palabras! La frase tiene {numero} palabras.')
    else:
      print('==============================================')
      print('===== 1. Adivine la letra ====================')
      print('===== 2. Adivine la palabra ==================')
      print('===== 3. Adivine la frase ====================')
      print('==============================================')
      eleccion = int(input('Seleccione una opción: '))
      if eleccion == 1:
        letra = input('Introduzca una letra: ')
        lista_intentos_letras.append(letra.lower())
        if letra.lower() in frase_adivinar.lower():
          for indice, letra_palabra in enumerate(frase_adivinar):
            if letra.lower() == letra_palabra.lower():
              frase_pista[indice] = letra.lower()
          ff = ('').join(frase_pista)
          print(f'Intento correcto. Existe la letra. {ff}. {numero_intentos} intentos disponibles.')
        else:
          numero_intentos -= 1
          ff = ('').join(frase_pista)
          print(f'Intento erróneo. No existe la letra. {ff}. {numero_intentos} intentos disponibles.')
      elif eleccion == 2:
        pal = input('Adivine la palabra: ')
        lista_intentos_palabras.append(pal.lower())
        if pal.lower() in frase_separada:
          inicio_pal = frase_adivinar.find(pal)
          for i, lp in enumerate(pal):
            frase_pista[inicio_pal+i] = lp
          ff = ('').join(frase_pista)
          print(f'Intento correcto. Existe la palabra. {ff}. {numero_intentos} intentos disponibles.')
        else:
          numero_intentos -= 1
          ff = ('').join(frase_pista)
          print(f'Intento erróneo. No existe la palabra. {ff}. {numero_intentos} intentos disponibles.')
      elif eleccion == 3:
        frase = input('Adivine la frase: ')
        lista_intentos_frases.append(frase.lower())
        if frase.lower() == frase_adivinar.lower():
          lin_f = (', ').join(lista_intentos_num)
          lil_f = (', ').join(lista_intentos_letras)
          lip_f = (', ').join(lista_intentos_palabras)
          lif_f = (', ').join(lista_intentos_frases)
          print('Has acertado!')
          print(f'Los intentos numero de letras fueron los siguientes: {lin_f}.')
          print(f'Los intentos de letras fueron los siguientes: {lil_f}.')
          print(f'Los intentos de palabras fueron los siguientes: {lip_f}.')
          print(f'Los intentos de frases fueron los siguientes: {lif_f}.')
          break
        else:
          numero_intentos -= 1
          print(f'Intento erróneo. La frase no es {frase}. {numero_intentos} intentos disponibles.')
      else:
        print('Error. Introduzca una opción correcta.')

```

**5) Adivina frase + puntos / comodines.**

Admin introduce una frase a adivinar, un número máximo de intentos para adivinarlo, una puntuación inicial, una puntuación por acierto, una puntuación por error y una puntuación por comodin.

Pide al usuario adivinar el número de palabras (ej1) en la frase. Si el usuario se equivoca, se le indica si el numero es mayor o menor y el número de intentos restantes así como los puntos restantes.

Pide al usuario si desea adivinar letras o palabras:
 

*   Letras: Se pide la letra. Muestra al usuario las letras que se van adivinando en la frase. Si se equivoca se resta 1 intento.
*   Palabras: Se pide la palabra. Muestra al usuario las palabras que se van adivinando en la frase. Si se equivoca se resta 1 intento.
*   Adivinar frase: Si se equivoca se resta 1 intento.
*   Comodin: Se muestra al usuario los valores únicos de letras pendientes de adivinar que hay en la frase y se le pide que letra desea desbloquear. Se desbloquean las letras. Se le resta la puntuación de comodín.
 
Si el usuario acierta una letra o palabra, se le suma la puntuación por acierto.

Si el usuario se equivoca una letra o palabra, se le resta la puntuación por error.

El programa termina cuando el usuario ha adivinado la frase, si se supera el máximo número de intentos o si se ha perdido todas las puntuaciones. Al final independientemente de haber adivinado o no, muestra los intentos insertados por tipo letra,palabra, frase y comodín.

```
import getpass

frase_adivinar = getpass.getpass('Introduzca la frase para jugar: ')

while True:
  print('==================================================')
  print('===== 1. Fácil ===================================')
  print('===== 2. Medio ===================================')
  print('===== 3. Difícil =================================')
  print('===== 4. Experto =================================')
  print('==================================================')
  d = int(input('Introduzca la dificultad: '))
  if d == 1:
    dificultad = 'fácil'
    puntuacion = 20
    puntuacion_por_acierto = 10
    puntuacion_por_error = 5
    num_comodines = 3
    break
  elif d == 2:
    dificultad = 'medio'
    puntuacion = 15
    puntuacion_por_acierto = 10
    puntuacion_por_error = 5
    num_comodines = 2
    break
  elif d == 3:
    dificultad = 'difícil'
    puntuacion = 11
    puntuacion_por_acierto = 10
    puntuacion_por_error = 5
    num_comodines = 1
    break
  elif d == 4:
    dificultad = 'experto'
    puntuacion = 10
    puntuacion_por_acierto = 10
    puntuacion_por_error = 9
    num_comodines = 0
    break
  else:
    print('No has elegido una dificultad válida.')

print(f'Empieza el juego. Nivel {dificultad}. Puntuacion inicial: {puntuacion}. Puntuacion por error: {puntuacion_por_error}. Puntuacion por acierto: {puntuacion_por_acierto}. Comodines: {num_comodines}.')

frase_separada = (frase_adivinar.lower()).split()
acertar_palabras = False
lista_intentos_num = []
lista_intentos_letras = []
lista_intentos_palabras = []
lista_intentos_frases = []
lista_intentos_comodines = []
frase_pista = []

for palabra in frase_separada:
  frase_pista.extend(['*'] * len(palabra))
  if palabra == frase_separada[-1]: break 
  else: frase_pista.append(' ')

while True:
  if puntuacion <= 0:
    print('Has perdido!')
    lin_f = (', ').join(lista_intentos_num)
    lil_f = (', ').join(lista_intentos_letras)
    lip_f = (', ').join(lista_intentos_palabras)
    lif_f = (', ').join(lista_intentos_frases)
    lic_f = (', ').join(lista_intentos_comodines)
    print(f'Los intentos numero de letras fueron los siguientes: {lin_f}.')
    print(f'Los intentos de letras fueron los siguientes: {lil_f}.')
    print(f'Los intentos de palabras fueron los siguientes: {lip_f}.')
    print(f'Los intentos de frases fueron los siguientes: {lif_f}.')
    print(f'Los intentos de comodines fueron los siguientes: {lic_f}.')
    break
  else:
    if not acertar_palabras:
      numero = int(input('Adivine el número de palabras: '))
      lista_intentos_num.append(str(numero))
      if numero > len(frase_separada):
        puntuacion -= puntuacion_por_error
        print(f'Intento erróneo. La frase tiene menos de {numero} palabras. Puntuación: {puntuacion}.')
      elif numero < len(frase_separada):
        puntuacion -= puntuacion_por_error
        print(f'Intento erróneo. La frase tiene más de {numero} palabras. Puntuación: {puntuacion}.')
      else:
        acertar_palabras = True
        puntuacion += puntuacion_por_acierto
        print(f'Ha acertado el número de palabras! La frase tiene {numero} palabras. Puntuación: {puntuacion}.')
    else:
      print('==============================================')
      print('===== 1. Adivine la letra ====================')
      print('===== 2. Adivine la palabra ==================')
      print('===== 3. Adivine la frase ====================')
      print('===== 4. Comodín =============================')
      print('==============================================')
      eleccion = int(input('Seleccione una opción: '))
      if eleccion == 1:
        letra = input('Introduzca una letra: ')
        lista_intentos_letras.append(letra.lower())
        if letra.lower() in frase_adivinar.lower():
          for indice, letra_palabra in enumerate(frase_adivinar):
            if letra.lower() == letra_palabra.lower():
              frase_pista[indice] = letra.lower()
          puntuacion += puntuacion_por_acierto
          ff = ('').join(frase_pista)
          print(f'Intento correcto. Existe la letra. {ff}. Puntuación: {puntuacion}.')
        else:
          puntuacion -= puntuacion_por_error
          ff = ('').join(frase_pista)
          print(f'Intento erróneo. No existe la letra. {ff}. Puntuación: {puntuacion}.')
      elif eleccion == 2:
        pal = input('Adivine la palabra: ')
        lista_intentos_palabras.append(pal.lower())
        if pal.lower() in frase_separada:
          inicio_pal = frase_adivinar.find(pal)
          for i, lp in enumerate(pal):
            frase_pista[inicio_pal+i] = lp
          puntuacion += puntuacion_por_acierto
          ff = ('').join(frase_pista)
          print(f'Intento correcto. Existe la palabra. {ff}. Puntuación: {puntuacion}.')
        else:
          puntuacion -= puntuacion_por_error
          ff = ('').join(frase_pista)
          print(f'Intento erróneo. No existe la palabra. {ff}. Puntuación: {puntuacion}.')
      elif eleccion == 3:
        frase = input('Adivine la frase: ')
        lista_intentos_frases.append(frase.lower())
        if frase.lower() == frase_adivinar.lower():
          lin_f = (', ').join(lista_intentos_num)
          lil_f = (', ').join(lista_intentos_letras)
          lip_f = (', ').join(lista_intentos_palabras)
          lif_f = (', ').join(lista_intentos_frases)
          lic_f = (', ').join(lista_intentos_comodines)
          puntuacion += puntuacion_por_acierto
          print(f'Has acertado! Puntuación final: {puntuacion}.')
          print(f'Los intentos numero de letras fueron los siguientes: {lin_f}.')
          print(f'Los intentos de letras fueron los siguientes: {lil_f}.')
          print(f'Los intentos de palabras fueron los siguientes: {lip_f}.')
          print(f'Los intentos de frases fueron los siguientes: {lif_f}.')
          print(f'Los intentos de comodines fueron los siguientes: {lic_f}.')
          break
        else:
          puntuacion -= puntuacion_por_error
          print(f'Intento erróneo. La frase no es {frase}. Puntuación: {puntuacion}.')
      elif eleccion == 4:
        if num_comodines == 0 and d == 4:
          print('Esta dificultad no tiene comodines!')
        elif num_comodines == 0:
          print('No te quedan comodines!')
        else:
          set_letras_frase = set()
          set_letras_usadas = set(lista_intentos_letras)
          for l in frase_adivinar:
            if l != (' '):
              set_letras_frase.add(l.lower())
          s = sorted(set_letras_frase.difference(set_letras_usadas))
          s_f = (', ').join(s)
          print(f'Estas son las letras que puede utilizar como comodín: {s_f}.')
          letra_comodin = input('Elija letra: ')
          lista_intentos_comodines.append(letra_comodin.lower())
          num_comodines -= 1
          if letra_comodin.lower() not in s_f:
            print(f'Letra errónea. Elija una letra correcta. Comodín perdido. {num_comodines} comodines disponibles.')
          else:
            for indice, letra_palabra in enumerate(frase_adivinar):
              if letra_comodin.lower() == letra_palabra.lower():
                frase_pista[indice] = letra_comodin.lower()
            ff = ('').join(frase_pista)
            print(f'Desvelando comodín {letra_comodin.lower()}... {ff}. {num_comodines} comodines disponibles.')
      else:
        print('Error. Introduzca una opción correcta.')
```
